```python
import pandas as pd
df = pd.read_csv('hh.zip', encoding='utf-8-sig')
print("Список всех колонок в датафрейме:")
print(df.columns.tolist())
```

```python
# Проверка и обработка region_name
print("\nАнализ region_name:")
print("Количество пропусков в region_name:", df['region_name'].isnull().sum())
print("Уникальные значения в region_name:")
print(df['region_name'].unique())

# Проверка и обработка region_district_name
print("\nАнализ region_district_name:")
print("Количество пропусков в region_district_name:", df['region_district_name'].isnull().sum())
print("Уникальные значения в region_district_name:")
print(df['region_district_name'].unique())

# Заполняем пропуски в region_district_name значением 'Не указан'
df['region_district_name'] = df['region_district_name'].fillna('Не указан')

print("\nПроверка после обработки region_district_name:")
print("Количество пропусков в region_district_name:", df['region_district_name'].isnull().sum())
```

```python
# Проверка и обработка soft_skills
print("\nАнализ soft_skills:")
print("Количество пропусков в soft_skills:", df['soft_skills'].isnull().sum())
print("Уникальные значения в soft_skills:")
print(df['soft_skills'].unique())

# Заполняем пропуски в soft_skills значением 'Не указаны'
df['soft_skills'] = df['soft_skills'].fillna('Не указаны')

print("\nПроверка после обработки soft_skills:")
print("Количество пропусков в soft_skills:", df['soft_skills'].isnull().sum())
```

```python
# Проверка и обработка experience_category
print("\nАнализ experience_category:")
print("Количество пропусков в experience_category:", df['experience_category'].isnull().sum())
print("Уникальные значения в experience_category:")
print(df['experience_category'].unique())

# Заполняем пропуски в experience_category значением 'Не указан'
df['experience_category'] = df['experience_category'].fillna('Не указан')

print("\nПроверка после обработки experience_category:")
print("Количество пропусков в experience_category:", df['experience_category'].isnull().sum())
```

```python
print("\nПроверка после обработки всех колонок:")
print("Количество пропусков в каждой колонке:")
print(df.isnull().sum())
```

```python
print("\nВсе значения в каждой колонке:")
for column in df.columns:
    print(f"\n{column}:")
    print(df[column].unique())
    # Анализ и обработка пропусков в каждой колонке
    print("\nАнализ пропусков в каждой колонке:")
    missing_stats = df.isnull().sum()
    missing_percentage = (missing_stats / len(df)) * 100
    
    missing_info = pd.DataFrame({
        'Количество пропусков': missing_stats,
        'Процент пропусков': missing_percentage
    })
    
    # Сортируем по количеству пропусков
    missing_info = missing_info.sort_values('Количество пропусков', ascending=False)
    
    # Выводим информацию о пропусках
    print(missing_info)
    
    # Обработка пропусков в числовых колонках
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numeric_columns:
        if df[col].isnull().sum() > 0:
            # Заполняем пропуски медианой
            df[col] = df[col].fillna(df[col].median())
            print(f"\nЗаполнены пропуски в числовой колонке {col} медианой")
    
    # Обработка пропусков в категориальных колонках
    categorical_columns = df.select_dtypes(include=['object']).columns
    for col in categorical_columns:
        if df[col].isnull().sum() > 0:
            # Заполняем пропуски модой (наиболее частым значением)
            mode_value = df[col].mode()[0]
            df[col] = df[col].fillna(mode_value)
            print(f"\nЗаполнены пропуски в категориальной колонке {col} значением '{mode_value}'")
    
    # Проверяем результат
    print("\nПроверка после обработки всех пропусков:")
    print("Количество пропусков в каждой колонке:")
    print(df.isnull().sum())
```
