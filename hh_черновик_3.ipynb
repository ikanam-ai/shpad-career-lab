```python
import pandas as pd
import numpy as np

def clean_and_filter_salaries_3(df):
    # Поверхностная фильтрация
    df_clean = df[
        (df['salary_to'] <= 2_000_000)].copy()
    
    
    def filter_group(group):
        # Отдельно сохраняем строки, где зарплата == 0
        zeros = group[(group['salary_from'] == 0) | (group['salary_to'] == 0)]
        non_zeros = group[(group['salary_from'] != 0) & (group['salary_to'] != 0)]
        # Для редких вакансий 
        if len(non_zeros) < 20:
            filtered = non_zeros[
                (non_zeros['salary_from'] >= 10_000) & 
                (non_zeros['salary_to'] <= 1_000_000)
            ]
        else: # Для остальных вакансий
            q1_from, q3_from = non_zeros['salary_from'].quantile([0.04, 0.96])
            q1_to, q3_to = non_zeros['salary_to'].quantile([0.04, 0.96])
            iqr_from = q3_from - q1_from
            iqr_to = q3_to - q1_to
            lower_bound_from = q1_from - 1.5 * iqr_from
            upper_bound_from = q3_from + 1.5 * iqr_from
            lower_bound_to = q1_to - 1.5 * iqr_to
            upper_bound_to = q3_to + 1.5 * iqr_to
            
            filtered = non_zeros[
                (non_zeros['salary_from'] >= lower_bound_from) & 
                (non_zeros['salary_from'] <= upper_bound_from) &
                (non_zeros['salary_to'] >= lower_bound_to) & 
                (non_zeros['salary_to'] <= upper_bound_to)
            ]

        # Объединяем отфильтрованные и нулевые строки
        return pd.concat([filtered, zeros], ignore_index=True)

    df_filtered = df_clean.groupby('role_name', group_keys=False).apply(filter_group)

    #  СПЕЦИАЛЬНЫЕ ПРАВИЛА ДЛЯ ОТДЕЛЬНЫХ ПРОФЕССИЙ 
    # Расширяем границы для стажёровок и подработки
    intern_mask = df_filtered['role_name'].str.contains(
        'Стажёр|Intern|Подработка|подработка|ПОДРАБОТКА',
        case=False,
        na=False
    )
    df_filtered.loc[intern_mask, 'salary_from'] = df_filtered.loc[intern_mask, 'salary_from'].clip(lower=5_000)
     # Расширяем границы для топ-менеджеров
    executive_mask = df_filtered['role_name'].str.contains(
        r'\b(CEO|CTO|Директор|Генеральный директор|Head of)\b',
        case=False,
        na=False
    )
    df_filtered.loc[executive_mask, 'salary_to'] = df_filtered.loc[executive_mask, 'salary_to'].clip(upper=2_000_000)


    return df_filtered
```

```python
df_clean_1=clean_and_filter_salaries_3(df_cleaned)
df_clean_1
```

```python
# Функция для вычисления медианы, игнорируя нулевые значения
def median_nonzero(series):
    nonzero = series[series != 0]
    if len(nonzero) == 0:
        return np.nan
    return nonzero.median()

grouped_1 = df_clean_1.groupby('role_name').agg(
    salary_from = ('salary_from', median_nonzero),
    salary_to = ('salary_to', median_nonzero),
    id_count = ('id', 'count')
).reset_index()


grouped = grouped_1.sort_values(by='salary_to', ascending=False)

grouped.head(10)
```

Отдельно считаем медиану по ненулевым salary_from и salary_to, считаем медиану по этим медианам и сортируем по убыванию 

```python
def median_of_two(row):
    if row['salary_from'] != 0 and row['salary_to'] != 0:
        return np.median([row['salary_from'], row['salary_to']])
    else:
        return np.nan

# Создаем новый столбец с медианой из двух зарплат
df_clean_1['salary_median'] = df_clean_1.apply(median_of_two, axis=1)

# Группируем и считаем медиану по этому новому столбцу
grouped = df_clean_1.groupby('role_name').agg(
    salary_median=('salary_median', 'median'),
    id_count=('id', 'count')
).reset_index()

# Сортируем по медиане зарплаты
grouped_sorted = grouped.sort_values(by='salary_median', ascending=False)

# Выводим топ-10
grouped_sorted.head(10)
```

Функция

```python
import pandas as pd
import re
from typing import Optional, Tuple, Union

def calculate_median_salaries(
    df: pd.DataFrame,
    professions: list[str],
    experience: str | None = None,
    regions: list[str] | None = None,
    employment_types: list[str] | None = None,
    is_open: bool | None = None,
) -> tuple[float, float] | None:

    
    filtered_df = df.copy()
    
    # 1. Фильтрация по специальности
    if professions:
        profession_regex = "|".join([re.escape(p) for p in professions])
        mask = (
            filtered_df["name"].str.contains(profession_regex, case=False, na=False, regex=True) |
            filtered_df["role_name"].str.contains(profession_regex, case=False, na=False, regex=True)
        )
        filtered_df = filtered_df[mask]
    
    # 2. Фильтрация по опыту работы
    if experience:
        filtered_df = filtered_df[filtered_df["experience"] == experience]
    
    # 3. Фильтрация по регионам
    if regions:
        filtered_df = filtered_df[filtered_df["region_name"].isin(regions)]
    
    # 4. Фильтрация по типу занятости
    if employment_types:
        filtered_df = filtered_df[filtered_df["employment_label"].isin(employment_types)]
    
    # 5. Фильтрация по статусу вакансии
    if is_open is not None:
        filtered_df = filtered_df[filtered_df["is_open"] == int(is_open)]
    
    # 6. Фильтрация положительных зарплат
    filtered_df = filtered_df[
        (filtered_df["salary_from"] > 0) & 
        (filtered_df["salary_to"] > 0)
    ]
    
    # Отладочная информация
    print(f"Найдено вакансий после всех фильтров: {len(filtered_df)}")
    if not filtered_df.empty:
        print("Примеры вакансий:")
        print(filtered_df[["name", "salary_from", "salary_to", "region_name", "is_open"]].head())
    
    if filtered_df.empty:
        return None
    
    # Расчет медиан только для положительных значений
    median_from = filtered_df["salary_from"].median()
    median_to = filtered_df["salary_to"].median()
    
    return (median_from, median_to)
```

```python
result = calculate_median_salaries(
    df=df_clean_1,
    professions=['Программист 1С'],
    experience="1-3",
    regions=["Москва"],
    employment_types=["Полная занятость"],
    is_open=False
)

if result:
    median_from, median_to = result
    print(f"Медиана salary_from: {median_from:.2f} руб.")
    print(f"Медиана salary_to: {median_to:.2f} руб.")
else:
    print("Подходящих вакансий с положительными зарплатами не найдено")
```
